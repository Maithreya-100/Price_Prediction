# cleaning

```python
#no need of area type,avaliability,society,balcony 
df2=df1.drop(["area_type","society","balcony","availability"],axis="columns")
df2
```

```python
df2.isnull().sum()
```

```python
#dropna is used to remove all null values
df3=df2.dropna()
df3.isnull().sum()
```

```python
df3.shape
```

```python
#since size contains string and different data types 
df3["bhk"]=df3["size"].apply(lambda x:x.split(" ")[0])
```

```python
df3
```

```python
df3["total_sqft"].unique()
```

```python
#sqft contains range values so clear them all
#define a fnction to convert all the values into integer
def val(x):
    tokens=x.split("-")
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df3["total_sqft"].unique()
```

```python
df3.index[df3["total_sqft"]=='1521'][0]
```

```python
df4=df3.copy()
df4["total_sqft"]=df3["total_sqft"].apply(val)
```

```python
df4
```

```python
#to verify whether total sqft contains None or not
if "None" in df4["total_sqft"].values:
    print("t")
else:
    print("F")
```

```python
#finding price per sqft
df5=df4.copy()
df5["price_per_sqft"]=df5["price"]*100000/df5["total_sqft"]
```

```python
df5
```

```python
#dimensionality curse
len(df5["location"].unique())
```

```python
df5.location=df5.location.apply(lambda x: x.strip())
```

```python
loc=df5.groupby("location")["location"].agg("count")
loc
```

```python
loc.sort_values(ascending=False)
```

```python
loc_stat=loc[loc<=10]
df5.location=df5.location.apply(lambda x: "other" if x in loc_stat else x)
```

```python
len(df5.location.unique())
```

```python
df5.head(10)
```

```python
#outlayer removal
#cond: sqft for bedroom should be less than threshold value
df5["bhk"] = pd.to_numeric(df5["bhk"])
df5[df5.total_sqft/df5.bhk<300]
```

```python
df5['bhk'] = df5['bhk'].astype(float)
```

```python
#removing the rows whose sqft value by considering bhk ie.,<300
df6=df5[~(df5.total_sqft/df5.bhk<300)]
```

```python
df6.shape
```

```python
#need to remove min max values for each location bcz min:267 is not possible and max:176470 is also considered, bcz its a generic model
df6.price_per_sqft.describe()
```

```python
#"Statistics"
#function to remove min and max values
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_pps_outliers(df6)
df7
```

# Model Building


```python
#hot Encoding
dummies=pd.get_dummies(df9.location)
dummies
```

```python
df10=pd.concat([df9,dummies.drop("other",axis="columns")],axis='columns')
```

```python
df11=df10.drop("location",axis="columns")
df11
```

```python
#for model building we are using split method
X= df11.drop(['price'],axis='columns')
X
```

```python
Y=df11.price
Y
```

```python
#Training the model using train test split
#random state will not choose ramdom rows other every time you run the program it will choose random rows
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
Y_test
```

```python
Y_train
```

```python
#applying linear regression 
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,Y_train)
lr_clf.score(X_test,Y_test) #return the accuracy score
```

```python
#k fold cross validation to measure the accuracy
# this will compare the predicted result and original value and gives accuracy level
#so the model is almost 80perc accurate
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X ,Y, cv=cv)
```

```python
#H: found best model(linearregression,decision tree,Lasso) using GridSeachcV
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    p = np.zeros(len(X.columns))
    p[0] = sqft
    p[1] = bath
    p[2] = bhk
    if loc_index >= 0:
        p[loc_index] = 1

    return lr_clf.predict([p])[0]
```
